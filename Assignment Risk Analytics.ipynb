```python
#Q7. Suppose we have a list a = [4,9,7,14,54,33,67,42]. Create a new list b which contains, all the odd elements of list a, 
#    multiplied by seven . (Using single line of code)

a = [4,9,7,14,54,33,67,42]
b = [7*i for i in a if i%2 == 1]
print(b)
```
